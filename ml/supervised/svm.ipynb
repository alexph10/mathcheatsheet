# SVM — companion notebook

Companion for [`svm.md`](./svm.md). Linear vs RBF SVM on 2D data; visualize margin, decision boundary, and support vectors.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_moons
rng = np.random.default_rng(0)
```

```python
# Linearly separable-ish dataset + a curved one
Xa = np.vstack([rng.multivariate_normal([-1.5, 0], [[1,0],[0,1]], 80),
                rng.multivariate_normal([ 1.5, 0], [[1,0],[0,1]], 80)])
ya = np.array([0]*80 + [1]*80)

Xb, yb = make_moons(n_samples=200, noise=0.15, random_state=0)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
for ax, (X, y, kernel, title) in zip(axes,
        [(Xa, ya, 'linear', 'Linear SVM'),
         (Xb, yb, 'rbf',    'RBF SVM (gamma=2)')]):
    clf = SVC(kernel=kernel, C=1.0, gamma=2 if kernel=='rbf' else 'scale').fit(X, y)
    xs = np.linspace(X[:,0].min()-1, X[:,0].max()+1, 300)
    ys = np.linspace(X[:,1].min()-1, X[:,1].max()+1, 300)
    gx, gy = np.meshgrid(xs, ys)
    Z = clf.decision_function(np.c_[gx.ravel(), gy.ravel()]).reshape(gx.shape)
    ax.contourf(gx, gy, Z, levels=20, cmap='RdBu_r', alpha=0.4)
    ax.contour(gx, gy, Z, levels=[-1, 0, 1], colors='k', linestyles=['--','-','--'])
    ax.scatter(X[y==0,0], X[y==0,1], c='blue', s=20)
    ax.scatter(X[y==1,0], X[y==1,1], c='red',  s=20)
    sv = clf.support_vectors_
    ax.scatter(sv[:,0], sv[:,1], facecolors='none', edgecolors='black', s=120, lw=1.5, label='support vectors')
    ax.set_title(f'{title}, {len(sv)} SVs'); ax.legend(loc='lower right'); ax.set_aspect('equal')
plt.show()
```
