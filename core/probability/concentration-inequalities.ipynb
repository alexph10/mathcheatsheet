# Concentration inequalities — companion notebook

Companion for [`concentration-inequalities.md`](./concentration-inequalities.md). For a sum of $n$ iid Bernoulli$(0.5)$, we plot Markov / Chebyshev / Hoeffding upper bounds against the exact Binomial tail.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
n = 100; p = 0.5
mu = p; sigma2 = p*(1-p)
# Threshold t s.t. P(|mean - mu| >= t)
ts = np.linspace(0.02, 0.5, 50)

exact = 2*stats.binom.cdf(np.floor(n*(mu - ts)), n, p)  # two-sided lower-tail; symmetric ok for p=0.5
cheby = sigma2 / (n * ts**2)
hoeff = 2 * np.exp(-2 * n * ts**2)  # bounded support b-a = 1
# Markov on |X - mu| can't be directly applied without bound; demo on X itself

fig, ax = plt.subplots(figsize=(8, 5))
ax.semilogy(ts, exact, 'k', lw=2, label='exact (Binomial two-sided)')
ax.semilogy(ts, np.clip(cheby, None, 1), label='Chebyshev')
ax.semilogy(ts, np.clip(hoeff, None, 1), label='Hoeffding')
ax.set_xlabel('deviation t'); ax.set_ylabel('P(|mean - 0.5| >= t)  (log)')
ax.set_title(f'n = {n} Bernoulli(0.5) tail vs upper bounds'); ax.legend(); ax.grid(True, which='both', alpha=0.3)
plt.show()
```

Hoeffding tracks the true tail much more tightly than Chebyshev once $t$ grows past a few hundredths. Chebyshev only needs the variance; Hoeffding needs bounded support.
