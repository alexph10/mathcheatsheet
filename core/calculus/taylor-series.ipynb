# Taylor series — companion notebook

Companion for [`taylor-series.md`](./taylor-series.md). We plot truncations of $\sin x$ and show how the truncation error grows away from the expansion point.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial
```

```python
def taylor_sin(x, n_terms):
    # Sum sin's Maclaurin series to (2n_terms-1)th degree
    s = np.zeros_like(x, dtype=float)
    for k in range(n_terms):
        s += (-1)**k * x**(2*k+1) / factorial(2*k+1)
    return s

xs = np.linspace(-3*np.pi, 3*np.pi, 800)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
ax1, ax2 = axes
ax1.plot(xs, np.sin(xs), 'k', lw=2, label='sin(x)')
for n in [1, 2, 3, 5, 8]:
    ax1.plot(xs, taylor_sin(xs, n), '--', label=f'degree {2*n-1}')
ax1.set_ylim(-2, 2); ax1.legend(); ax1.set_title('Maclaurin truncations of sin(x)')
ax1.axvline(0, color='gray', lw=0.5); ax1.axhline(0, color='gray', lw=0.5)

for n in [1, 2, 3, 5, 8]:
    err = np.abs(np.sin(xs) - taylor_sin(xs, n))
    ax2.semilogy(xs, err + 1e-16, label=f'degree {2*n-1}')
ax2.set_title('Absolute truncation error |sin(x) - T_n(x)|'); ax2.legend()
ax2.set_ylim(1e-16, 1e3)
plt.show()
```

The truncation error grows polynomially in $|x|$ but the polynomial degree grows linearly with the number of terms — so adding terms helps locally but the radius of usefulness for any finite truncation is bounded.
