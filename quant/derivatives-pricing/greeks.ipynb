# Greeks — companion notebook

Companion for [`greeks.md`](./greeks.md). Plot Delta, Gamma, and Vega for a Black-Scholes call across spot at several maturities.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def bs_d1d2(S, K, r, sigma, T):
    d1 = (np.log(S/K) + (r + 0.5*sigma**2)*T) / (sigma*np.sqrt(T))
    return d1, d1 - sigma*np.sqrt(T)
def call_price(S, K, r, sigma, T):
    d1, d2 = bs_d1d2(S, K, r, sigma, T)
    return S*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2)
def delta(S, K, r, sigma, T):
    d1, _ = bs_d1d2(S, K, r, sigma, T); return norm.cdf(d1)
def gamma(S, K, r, sigma, T):
    d1, _ = bs_d1d2(S, K, r, sigma, T); return norm.pdf(d1) / (S*sigma*np.sqrt(T))
def vega(S, K, r, sigma, T):
    d1, _ = bs_d1d2(S, K, r, sigma, T); return S*norm.pdf(d1)*np.sqrt(T)
```

```python
K, r, sigma = 100.0, 0.03, 0.20
S = np.linspace(60, 140, 400)
Ts = [1.0, 0.5, 0.1]

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for T in Ts:
    axes[0].plot(S, delta(S, K, r, sigma, T), label=f'T={T}')
    axes[1].plot(S, gamma(S, K, r, sigma, T), label=f'T={T}')
    axes[2].plot(S, vega (S, K, r, sigma, T), label=f'T={T}')
for ax, title in zip(axes, ['Delta', 'Gamma', 'Vega']):
    ax.axvline(K, color='gray', ls='--', alpha=0.6)
    ax.set_xlabel('Spot S'); ax.set_title(title); ax.legend()
plt.tight_layout(); plt.show()
```
